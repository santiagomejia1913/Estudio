#Arquitectura de computo en GCP

- Compute Engine (Maquinas virtuales): Vienen en uno de dos sabores, Linux (Devian, Ubuntu, red hat, etc) o servidores windows. Rara vez se ven servidores de Mac. 

- GKE (Google Kubernets Engine): Se piensa en contenedores. Orquestrar diferentes tipos de contenedores que pueden generar un cluster para que todos los contenedores (Ya sean propios desarrollados {in-house} o a traves de compras de software a terceros) se puedan instalar directamente en nuestro cluster. En palabras sencillas un contenedor es un ejecutable para la nube. Estos ejecutables se pueden correr en Kubernet engine donde se tiene un monton de flexibilidad en como customizar y como configurar las maquinas y el cluster que estamos utilizando. 

Nota: 
A partir de acá entramos mas en el mundo de Serverless Platform: Permite enfocar en el desarrollo de la aplicación y no en el desarrollo de la infraestructura. 

- App Engine: Normalmente se utiliza para poder correr aplicaciones Web (Flask, Django). 

- Cloud Run: Es lo mismo que que App Engine pero esta enfocada en contenedores. Te permite agarrar un contenedor, indicar que lo corra y Cloud Run automaticamente lo corre. No se necesita preocupar por instalar software previo, por ver que es lo que va a correr este contenedor, ni por instalar Kubernetes, ni instalar Docker. Solo se indica, que se tienen un contenedor en un registry, para que lo baje y lo comience a correr.

- Cloud Functions: Solo preocuparse por la logica aplicativa que queremos utilizar y ya esta completamente administrada. Es decir, solo es darle codigo para que lo corra, mientras que en "Cloud Run" se debe pasar un contenedor que tiene el codigo y se debe preocupar por las dependencias del sistema operativo.

- Firebase: Es la automatización completa de la nube para solo preocuparse por el Front-end. Es decir, es una forma de automatizar todo el back-end.

# Regiones y zonas en GCP

La región es una zona geográfica donde se encuentran varias zonas separadas por una distancia determinada, en las cuales cada una de estas zonas tienen adentro los data centers.

Muchas veces vamos a requerir que nuestra aplicación o software este en varias
zonas y/o regiones por alta disponibilidad y con latencia, para evitar que una catástrofe ya sea natural o humana altere nuestros servicios.

Los puntos de presencia son data centers de los proveedores de servicio de internet locales con los que Google tiene contratos y/o convenios

gcping Es una excelente herramienta que nos permite medir la latencia desde nuestro punto a las regiones de GCP.
