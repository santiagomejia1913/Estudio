Un decorador es una funci칩n que recibe como par치metro otra funci칩n, le a침ade cosas y retorna una funci칩n diferente. Tienen la misma estructura que los Closures pero en vez de variables lo que se env칤a es una funci칩n.

Ejemplo 1

```python
def funcion_decoradora(funcion_parametro):

    def funcion_interior():

        #Acciones adicionales que decoran

        print("Vamos a realizar un c치lculo: ")

        funcion_parametro()

        #Acciones adicionales que decoran

        print("Hemos terminado el c치lculo")

    return funcion_interior


# Para llamar la funci칩n decoradora se usa el car치cter "@" y a continuaci칩n el nombre de esta.
# Python entiende que la funci칩n adyacente hacia abajo es la funci칩n a decorar y la toma como parametro.

@funcion_decoradora
def suma():
    print(15+20)

suma()
```

Ejemplo 2

```python

def decorador(func):
    def envoltura():
        print('Esto se a침ade a mi funcion original')
        func()
    return envoltura

def mayusculas(func):
    def envoltura(texto):
        return func(texto).upper()
    return envoltura


@decorador
def saludo():
    print('Hola')

@mayusculas
def mensaje(nombre):
    return f'{nombre}, recibistes un mensaje'

saludo()
print(mensaje('duvan'))
```

##8.Iteradores

Antes de entender qu칠 son los iteradores, primero debemos entender a los iterables.

Iterables: Son todos aquellos objetos que podemos recorrer en un ciclo. Son aquellas estructuras de datos divisibles en elementos 칰nicos que yo puedo recorrer en un ciclo.

Cuando hacemos un ciclo, python no est치 iterando(internamente). Lo que sucede es que ese iterable se convierte en un objeto especial el cual se llama iterador.

```python
# Creando un iterador

my_list = [1,2,3,4,5]
my_iter = iter(my_list)

# Iterando un iterador

print(next(my_iter))

# Cuando no quedan datos, la excepci칩n StopIteration es elevada
```

- iter convierte un iterable en un iterador.
- Una vez que se hace, se usa al next() para ir al siguiente elemento del iterador
- Cuando ya no quede ning칰n elemento python elevara una excepci칩n "StopIteration"

```python
# Creando un iterador

my_list = [1,2,3,4,5]
my_iter = iter(my_list)

# Iterando un iterador

while True: #ciclo infinito
  try:
    element = next(my_iter)
    print(element)
  except StopIteration:
    break
```

Momento impactante: El ciclo 랁or dentro de Python, no existe. Es un while con StopIteration. 游뱚游뱚游뱚

Es decir, se un alias de un ciclo While infinito con un Try except que controla el Error StopIteration

La estructura For por lo tanto es "azucar sintactico"

```python
class EvenNumber:
    
    """
    Clase que implementa un iterador de todos los n칰meros pares, o los n칰meros pares hasta un m치ximo."""

def __init__(self,max=None):
    self.max = max
def __iter__(self):
    self.num = 0
    return self

def __next__(self):
    if not self.max or self.num <= self.max:
        result= self.num
        self.num += 2
        return result

    else:
        raise StopIteration
```

##9.Generadores

Es el "Sugar Syntax" de los iteradores.

Los generadores son funciones que guardan un estado. Es un iterador escrito de forma m치s simple.

```python
def my_gen():

  """un ejemplo de generadores"""

  print('Hello world!')
  n = 0
  yield n # es exactamente lo mismo que return pero detiene la funci칩n, 
  #cuando se vuelva a llamar a la funci칩n, seguir치 desde donde se qued칩

  print('Hello heaven!')
  n = 1
  yield n

  print('Hello hell!')
  n = 2
  yield n


a = my_gen()
print(next(a)) # Hello world!
print(next(a)) # Hello heaven!
print(next(a)) # Hello hell!
#print(next(a))
```

**Generator Expression**



Es como list comprehension pero mucho mejor, porque podemos manejar mucha cantidad
de informaci칩n sin tener problemas de rendimiento

```python
#Generator expression

my_list = [0,1,4,7,9,10]

my_second_list = [x*2 for x in my_list] #List comprehension
my_second_gen = (x*2 for x in my_list) #Generator expression
```

```python
for element in my_second_gen:
  print(element)
```

Un 랇ist comprehension es una manera sencilla de implementar un ciclo el cual nos permite crear una nueva lista con elementos basados en los valores de una lista existente. Dicha lista creada almacena cada uno de estos nuevos valores.

A diferencia de un List comprehension que puede ocupar mucha memoria pues cada valor es almacenado en la variable, un Generator expression solo los recorre, sin guardarlos en una lista nueva. Permite traer un elemento a la vez cuando se recorra usando un ciclo for.

Como se puede ver en el ejemplo, la diferencia en la sintaxis 칰nicamente es en el uso de par칠ntesis en lugar del uso de corchetes.

Ventajas de los Generadores:

- Es mas f치cil de escribir que un iterador
- Ahorra Tiempo y Memoria
- Permite guardar secuencias infinitas

##10.Sets

Los sets son una estructura de datos muy similares a las listas en cuanto a su forma, pero presentan ciertas caracter칤sticas particulares:

- Los sets son conjuntos con elementos inmutables (Por lo que por ejemplo una lista no podria estar contenido en un Set)

- Cada elemento del set es 칰nico, esto es que no se admiten duplicados, aun si durante la definici칩n del set se agregan elementos repetidos pyhton solo guarda un elemento

- los sets guardan los elementos en desorden

Para declararlos se utilizan los {} parecido a los diccionarios solo que carece de la composici칩n de conjunto {a:b, c:d}

```python
# set de enteros
my_set = {1, 3, 5}
print(my_set)

# set de diferentes tipos de datos
my_set = {1.0, "Hi", (1, 4, 7)}
print(my_set)
```

Los sets no pueden ser le칤dos como las listas o recorridos a trav칠s de slices, esto debido a que no tienen un criterio de orden. Sin embargo si podemos agregar o eliminar items de los sets utilizando m칠todos:

- add(): nos permite agregar elementos al set, si se intenta agregar un elemento existente simplemente python los ignorara
update(): nos permite agregar m칰ltiples elementos al set
- remove(): permite eliminar un elemento del set, en el caso en que no se encuentre presente dicho elemento, Python elevar치 un error
- discard(): permite eliminar un elemento del set, en el caso en que no se encuentre presente dicho elemento, Python dejar치 el set intacto, sin elevar ning칰n error.
- pop(): permite eliminar un elemento del set, pero lo har치 de forma aleatoria.
- clear(): Limpia completamente el set, dej치ndolo vaci칩.

```python
#ejemplo de operaciones sobre sets 
my_set = {1, 2, 3} 
print(my_set) #Output {1, 2, 3} 

#a침adiendo un elemento al set 
my_set.add(4) 
print(my_set) #Output {1, 2, 3, 4}

#a침adiendo varios elementos al set, python ignorar치 elementos repetidos 
my_set.update([1, 5, 6]) 
print(my_set) #Output {1, 2, 3, 4, 5, 6}

# eliminado elementos del set 
my_set.discard(1) 
print(my_set) #Output {2, 3, 4, 5, 6}

my_set.discard(4) 
print(my_set) #Output {2, 3, 5, 6}

# borrando un elemento aleatorio 
my_set.pop()
print(my_set) #Output el set menos un elemento aleatorio 

#limpiar el set 
my_set.clear()
print(my_set) # Output set()
```

- .discard() = Se indica como par치metro el elemento a eliminar, en caso de no exisitir el elemento no realiza ninguna operaci칩n y contin칰a con la ejecuci칩n.
- .remove() = Tambi칠n se indica como par치metro el elemento a eliminar, pero en este caso si no exististe el elemento arroja un error (KeyError) y detiene la ejecuci칩n.

Podemos utilizar estructuras de datos existentes para transformarlas a sets utilizando el m칠todo set:

```python
#usando listas para crear sets
my_list = [1, 2, 3, 3, 4, 5]
my_set = set(my_list)
print(my_set)  #output {1, 2, 3, 4, 5}

#usando tuplas para crear sets 
my_tuple = ('hola', 'hola', 1, 2)
my_set2 = set(my_tuple)
print(my_set2) #Output {'hola', 1}
```

**Operaciones de conjuntos**

Uni칩n

Resultado de juntar todos los elementos que tienen ambos, combinar ambos sets

```python
set_1 = {1, 2, 3}
set_2 = {3, 4, 5}
set_3 = set_1 | set_2 #  -> {1, 2, 3, 4, 5}

# O de forma mas explicita
set_1.union(set_2) #  -> {1, 2, 3, 4, 5}
```

Intersecci칩n

Los que se repiten en ambos sets

```python
set_1 = {1, 2, 3}
set_2 = {3, 4, 5}
set_3 = set_1 & set_2 #  -> {3}

# O de forma mas explicita
set_1.intersection(set_2) #  -> {3}
```

Diferencia

Tomar dos sets y de uno quitarle todos los elementos que tiene el otro

```python
set_1 = {1, 2, 3}
set_2 = {3, 4, 5}
set_3 = set_1 - set_2 #  -> {1, 2}

# O de forma mas explicita
set_1.difference(set_2) #  -> {1, 2}
```

Diferencia Sim칠trica

Tomar todos los elementos exceptos los que se repiten

```python
set_1 = {1, 2, 3}
set_2 = {3, 4, 5}
set_3 = set_1 ^ set_2 #  -> {1, 2, 4, 5}

# O de forma mas explicita
set_1.symmetric_difference(set_2) #  -> {1, 2, 4, 5}
```
