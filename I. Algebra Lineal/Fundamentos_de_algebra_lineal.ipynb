### Propiedades de las matrices

La multiplicacin de matrices es asociativa y distributiva, no es conmutativa

El producto interno de **++matrices ++**es:

- Asociativa: S Ax(BxC) = (AxB) x C
- Distributiva: S Ax(B+C) = (AxB) x (AxC)
- Conmutativa: NO BxC = CxB

El producto interno de **++vectores ++**es:

- Asociativa: S Ax(BxC) = (AxB) x C
- Distributiva: S Ax(B+C) = (AxB) x (AxC)
- Conmutativa: S BxC = CxB

## Sistemas de Ecuaciones Lineales

### Ejemplos de sistemas sin solucin, con una solucin y con infinitas soluciones

Tipos de sistemas de ecuaciones:

- Incompatible o sobredeterminado: Cuando no hay soluciones para el sistema. Hay ms ecuaciones que incgnitas
- Compatible determinado: Cuando el sistema tiene una nica solucin. Hay tantas ecuaciones como incgnitas (convergen en un slo punto)
- Compatible indeterminado: Cuando hay infinitas soluciones. Hay menos ecuaciones que incgnitas.

### Funcion Auxiliar para graficar vectores

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

v1 = np.array([2,5])
v2 = np.array([3,2])

# vecs: vectores.
# cols: colores.
# alpha: valor de transparencia.
def graficarVectores(vecs, cols, alpha=1):
    plt.figure()
    plt.axvline(x=0, color="grey", zorder=0)
    plt.axhline(y=0, color="grey", zorder=0)
    
    for i in range(len(vecs)):
        # El origen de los vectores inicia en el punto (0,0)
        x = np.concatenate([[0,0], vecs[i]])
        plt.quiver([x[0]],
                  [x[1]],
                  [x[2]],
                  [x[3]],
                  angles='xy', scale_units='xy', scale=1, 
                  color=cols[i], alpha=alpha)

graficarVectores([v1,v2], ['orange', 'blue'])
# Definimos los limites
plt.xlim(-1, 8)
plt.ylim(-1, 8)
```

### Espacio y Subespacio

```python
#%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
```

```python
v1 = np.array([1, 1])
v2 = np.array([-1, -1])
```

```python
for a in range(-10,10):
    for b in range(-10,10):
        plt.scatter(v1[0]*a + v2[0]*b, v1[1]*a + v2[1]*b,
                   marker = '.',
                   color = "orange")
        
plt.xlim(-25,25)
plt.ylim(-25,25)

plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')

plt.show()
```

```python
v1 = np.array([1, 0])
v2 = np.array([2, -3])
```

```python
for a in range(-10,10):
    for b in range(-10,10):
        plt.scatter(v1[0]*a + v2[0]*b, v1[1]*a + v2[1]*b,
                   marker = '.',
                   color = "orange")
        
plt.xlim(-25,25)
plt.ylim(-25,25)

plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')

plt.show()
```

La razn por la que los vectores (-1,-1) y (1,1) no pueden generar a R2 (teniendo en cuenta que la generacin del espacio se obtiene con la variacin de los escalares como lo hemos venido haciendo) es porque estos dos vectores no son linealmente independientes entre s, en trminos ms sencillos vase que estos vectores son multiplos entre ellos, basta con multiplicar por el escalar -1 a cualquiera de los dos para obtener el otro.

Estos vectores por tanto van a pertenecer al mismo espacio, en particular, generan un subespacio de R2 que sera: todos los vectores pertenecientes a R2 tal que la segunda coordenada sea igual a la primera, y = x. Para poder generar el espacio R2 (o en general Rn), se necesitan tomar n vectores que sean linealmente independientes entre ellos, que ninguno sea combinacin lineal de los otros.

Es por ello que en el segundo ejemplo cuando toma (1, 0) y (2, -3), esto genera todo el espacio R2, pues son dos vectores que son linealmente independientes, es decir, no existe un escalar tal que multiplicando a uno me permita obtener al otro.

```python
from mpl_toolkits.mplot3d import Axes3D

v1 = np.array([1, 0, 0])
v2 = np.array([2, -3, 0])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for a in range(-10,10):
    for b in range(-10,10):
        ax.scatter(v1[0]*a + v2[0]*b, v1[1]*a + v2[1]*b,
                   v1[2]*a + v2[2]*b,
                   marker = '.',
                   color = "orange")
        
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')

plt.show()
```

R2 es un subespacio de R3. Un hiperplano es una dimension menos de la que estamos trabajando. Si estamos trabajando en R2, un hiperplano es la recta.

## Normas

### Definicin y propiedades

Propiedades:
- Siempre es mayor a Cero
- Si la norma es cero, entonces esta en el origen
- Desigualdad triangular: Si tenemos dos vectores y los sumamos, la norma del vector resultante es menor o igual a la suma de las normas de los dos vectores originales.
- La norma de un escalar por un vector es igual a la multiplicacin del valor abosluto de ambos --> norma(k*v) = abs(k) * norma(v).

### Tipos de Normas

La norma se usa en machine learning para encontrar errores.
- La norma L0: devuelve la cantidad de elementos diferentes de 0 del vector.
- La norma L1: devuelve la suma de los valores absolutos de los elementos del vector.
- La norma L2: La magnitud del vector desde su origen (Euclidiana).

Nota: En machine learning se usa mucho la norma L2 pero elevada al cuadrado es decir, L2^2. La ventaja de esto es computacional. Podramos reunir muchos vectores en una matriz X, calcular X.dot(x_t) y tendramos all el cuadrado de las normas de todos esos vectores.

- La norma infinito: devuelve el valor absoluto ms grande del vector

image.png

Un aporte adicional que nohe visto se explique de forma amplia en Platzi, las normas se usan para determinar el modulo de regularizacin en la construccin de un modelo. La regularizacin se usa para castigar la complejidad de un modelo y evitar overfitting, las ms comunes son L2 para simplicidad y L1 para esparcidad. Para ms informacin se puede consultar aca en el crash course de Machine Learning de Google:

### Matrices Ortogonales y sus propiedades

Matriz Ortogonal

Es la matriz cuyos vectores filas son ortonormales mutuamente (lo mismo con los vectores columnas)
Cumplen con la siguiente identidad:
- A_tA = AA_t = I
- A_t = inv(A)

Nota:
Cuidado con los errores que podemos propagar al operar con las aproximaciones que hace python

```python
matriz = np.array([[1,0,0],
                   [0,1,0],
                   [0,0,1]
                  ])
print(matriz)
```

Comprobemos que las columnas son ortogonales entre las columnas

```python
print(matriz[:,0].dot(matriz[:,1]))
```

```python
print(matriz[:,0].dot(matriz[:,2]))
```

```python
print(matriz[:,1].dot(matriz[:,2]))
```

Comprobemos que la norma de las columnas de cada columna es 1

```python
print(np.linalg.norm(matriz[:,0]))
print(np.linalg.norm(matriz[:,1]))
print(np.linalg.norm(matriz[:,2]))
```

Como las columnas son ortogonales entre si y su norma es 1, entonces las columnas de la matriz son ortonormales.

Ahora verifiquemos que las filas cumplen con las condiciones anteriores para poder decir que la matriz es ortogonal.

```python
print(matriz[0,:].dot(matriz[1,:]))
print(matriz[0,:].dot(matriz[2,:]))
print(matriz[1,:].dot(matriz[2,:]))
```

```python
print(np.linalg.norm(matriz[0,:]))
print(np.linalg.norm(matriz[1,:]))
print(np.linalg.norm(matriz[2,:]))
```

Como las filas son ortogonales entre si y su norma es 1, entonces las filas de la matriz son ortonormales.

Por lo tanto como las filas y las columnas de la matriz son ortonormales, la matriz se considera ortogonal (Recordemos que el concepto de matriz ortonormal no existe).

## Determinante y traza

```python
import numpy as np
import matplotlib.pyplot as plt
```

Determinante

El determinante de una matriz nos habla sobre la transformacin que ejerce esa matriz sobre el espacio que est transformando.

Si estamos en R2 y el determinante es negativo, lo que hace es darnos el espejo de nuestro espacio.

Hablando en R2, si una determinante es negativa nos indica que la trasformacin rot el espacio 180 grado.

```python
def graficarVectores(vecs, cols, alpha=1):
    plt.figure()
    plt.axvline(x=0, color="grey", zorder=0)
    plt.axhline(y=0, color="grey", zorder=0)
    
    for i in range(len(vecs)):
        # El origen de los vectores inicia en el punto (0,0)
        x = np.concatenate([[0,0], vecs[i]])
        plt.quiver([x[0]],
                  [x[1]],
                  [x[2]],
                  [x[3]],
                  angles='xy', scale_units='xy', scale=1, 
                  color=cols[i], alpha=alpha)
```

```python
v1 = np.array([0,1])
v2 = np.array([1,0])

graficarVectores([v1,v2], ["blue","red"])

plt.xlim(-0.25, 2)
plt.ylim(-0.25, 2)

plt.show()
```

Con estos dos vectores podemos generar todo R2. Pero cuando los transformamos que ocurre? La transformacion se realiza con el producto una matriz.

```python
A = np.array([[2,0],[0,2]])
print(A)
```

```python
v1_transformado = A.dot(v1)
v2_transformado = A.dot(v2)

graficarVectores([v1_transformado,v2_transformado], ["orange","green"])

plt.xlim(-0.25, 3)
plt.ylim(-0.25, 3)

plt.show()
```

image.png

El determinante nos devuelve un numero, este es el rea que se forma con los vectores.

```python
det_A = np.linalg.det(A)

print(det_A)
```

```python
area_transformada = abs((v1_transformado[0]-v2_transformado[0]))*abs((v1_transformado[1]-v2_transformado[1]))
print(area_transformada)
```

Haciendo lo mismos con una matriz con determinante negativo

```python
B = A * [-1, 1]
print(B)
```

```python
det_B = np.linalg.det(B)
print(det_B)
```

```python
v1 = np.array([0,1])
v2 = np.array([1,0])

v1_transformado = B.dot(v1)
v2_transformado = B.dot(v2)

graficarVectores([v1_transformado, v2_transformado, v1, v2],
                ['blue','red','orange','green'])

plt.xlim(-3, 1.5)
plt.ylim(-0.5, 2.5)

plt.show()
```

Lo que esta ocurriendo al tener un signo negativo en el determinante es que nuestro espacio roto.

Entonces lo que nuestra matriz hizo ademas de amplificar nuestro vector fue invertir en este caso una de nuestras coordenadas, la espejo.

La matriz A con determinante 4 y la matriz B con determinante -4, modificarn nuestro espacio, en este caso lo amplificaron. La diferencia es que la matriz A lo amplifico sobre el mismo espacio y la matriz B lo amplific pero modifico su espacio y lo espejo.

La informacin del determinante es muy importante, pero que no nos deja saber todo lo que est ocurriendo con el espacio. No podemos saber si esta rotacin que est ocurriendo lo hace sobre el eje x o el eje y.
