## 1.Series y DataFrames en Pandas

Pandas maneja 2 objetos principales: 
- Pandas Series: Parecido a una array unidimensional que usa numpy, porque tambien es indexado y tambien puedo hacer slicing, puedo entrar directamente a los objetos que quiero buscar, resiste muchas operaciones matematicas y maneja muchos tipos de datos. 
- Pandas Dataframe: Parecido a las estructuras matriciales que vimos en numpy. Es un objeto hecho de arrays pero en una estructura matricial. Las filas y las columnas tienen indices, los indices de las columnas son sus nombres y los indices de las filas podemos asignarlos manualmente o toman un valor muy parecido al indexado que teniamos en numpy de 0 hasta donde llegue el dato final. Se pueden hacer slicing, operaciones matematicas, tiene tamaño variable y maneja distintos tipos de datos.

MERGE

```python
izq = pd.DataFrame({'key':['k0','k1','k2','k3'],
 'A':['A0', 'A1', 'A2', 'A3'],
 'B':['B0', 'B1', 'B2', 'B3']})

der = pd.DataFrame({'key':['k0','k1','k2','k3'],
 'C':['C0', 'C1', 'C2', 'C3'],
 'D':['D0', 'D1', 'D2', 'D3']})
```

```python
izq
```

```python
der
```

```python
izq.merge(der, on='key')
```

```python
izq = pd.DataFrame({'key':['k0','k1','k2','k3'],
 'A':['A0', 'A1', 'A2', 'A3'],
 'B':['B0', 'B1', 'B2', 'B3']})

der = pd.DataFrame({'key_2':['k0','k1','k2','k3'],
 'C':['C0', 'C1', 'C2', 'C3'],
 'D':['D0', 'D1', 'D2', 'D3']})
```

```python
izq.merge(der, left_on='key', right_on='key_2')
```

```python
izq = pd.DataFrame({'key':['k0','k1','k2','k3'],
 'A':['A0', 'A1', 'A2', 'A3'],
 'B':['B0', 'B1', 'B2', 'B3']})

der = pd.DataFrame({'key_2':['k0','k1','k2',np.nan],
 'C':['C0', 'C1', 'C2', 'C3'],
 'D':['D0', 'D1', 'D2', 'D3']})
```

```python
der
```

```python
izq.merge(der, left_on='key', right_on='key_2') #El valor por defecto es 'inner' Join
```

```python
izq.merge(der, left_on='key', right_on='key_2', how='left')
```

```python
izq.merge(der, left_on='key', right_on='key_2', how='right')
```

JOIN

Los Join hacen la union de dataframes con los indices del mismo, y no con las columnas como pasaba con el merge.

```python
izq=pd.DataFrame({'A':['A0', 'A1', 'A2'],
 'B':['B0', 'B1', 'B2']},
 index=['k0', 'k1', 'k2'])
```

```python
der = pd.DataFrame({'C':['C0', 'C1', 'C2'],
 'D':['D0', 'D1', 'D2']},
 index=['k0', 'k2', 'k3'])
```

```python
der
```

```python
izq.join(der, how='left')
```

```python
izq.join(der, how='right')
```

```python
izq.join(der, how='inner')
```

```python
izq.join(der, how='outer')
```

MELT

El método melt toma las columnas del DataFrame y las pasa a filas, con dos nuevas columnas para especificar la antigua columna y el valor que traía.

Por ejemplo, simplemente al imprimir las cinco primeras filas del DataFrame con las columnas de Name y Genre se tiene este resultado.

```python
df_books[['Name','Genre']].head(5)
```

```python
df_books[['Name','Genre']].head(5).melt()
```

Ahora cada resultado de las dos columnas pasa a una fila de este modo a tipo llave:valor.

```python
df_books.melt(id_vars='Year',value_vars='Genre')
```

Simplemente, podemos seleccionar las columnas que no quiero hacer melt usando el parámetro id_vars. Para este caso Year y también la única columna que quiero aplicar el melt, para este caso Genre con la propiedad value_vars.
