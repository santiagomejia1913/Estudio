## Limite

### Definición

El límite es un análisis de tendencia respecto a un punto. Es decir, hacia donde va el valor de nuestra función a medida que “x” se acerca al valor “a”, pero sin tocar nunca ese valor “a”. Por eso se llama límite, porque “tenemos prohibido tocar” ese valor “a”.

El concepto de limite es simplemente decir “Ya que no podemos usar cierto valor porque nos da un resultado indefinido, vamos a acercarnos mucho a ese cierto valor y ver que resultado nos da”.

### Para que son los limites laterales?

Decimos que si el limite por izquierda y por derecha son iguales cuando tienden a *a*, entonces el limite L existe.

Recordar: Los limites NO son evaluar funciones, sino saber que pasa con la función cuando se aproxima a cierto valor

## Derivada

La derivada de una función f(x) es otra función f '(x) que representa las pendientes de las rectas tangentes para cualquier punto de f(x). En consecuencia la derivada expresa la razón de cambio de una función.

## Implementación de la derivada discreta

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return x**2

x = np.linspace(-10,10, num = 1000)

def df(f):
  h = 0.000001
  return (f(x+h)-f(x))/h


y = f(x)

plt.plot(x,y, 'b')
plt.plot(x, df(f), 'r')
plt.grid()
```

## Maximos y Minimos y su aplicación en optimización

Es necesario tener en cuenta la optimización buscando una función objetivo, esto significa que, con ayuda de las derivadas y los límites podrémos encontrar el área más óptima para el desarrollo de las “oficinas”

para esto se hace lo siguiente:

**Paso 01.** entender el área y perimetro del espacio:

A = x * y
P = x + y + x = 2x + y

**Paso 02:** Tener un área, en este caso de muro, como restricción (50ml de perimetro en este caso)

2x + y = 50

**Paso 03:** Buscar la función objetivo, en este caso y (es la función que queremos optimizar)

y = 50-2x

**Paso 04:** Reescribir en terminos de la dependiente

área en terminos de x

A(x) = x(50-2x)
A(x) = -2x² + 50x

**Paso 05:** optimizar buscando su derivada e igualar a 0 (se iguala porque queremos saber en qué punto la x está en el punto más alto o más bajo)

A’(x) = -4x + 50 = 0

-4x + 50 = 0

x = -50/ -4 = 25/2 (mínima expresión)

Paso 06: Buscar datos a la izquierda y a la derecha evaluando el valor de x en la derivada

A’(x) = -4x + 50

a la izquierda:

A’(12.45) = -4(12.45) + 50 = 1/5 (pendiente positiva)

a la derecha

A’(12.55) = -4(12.55) + 50 = -1/5 (pendiente negativa)

Como a la izquierda la pendiente es positiva y a la derecha la pendiente es negativa, estamos hablando de un Maximo!!
