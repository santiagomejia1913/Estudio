# PROGRAMACIÓN ORIENTADA A OBJETOS

Que resuelve la programación orientada a objetos (vs. Programación estructurada)?
1. Codigo muy largo. 
2. Si algo falla, todo se rompe. 
3. Dificil de mantener.

Orientación a objetos

Surge a partir de los problemas qe tenemos y necesitamos plasmar en codigo (Observar los problemas en forma de objetos).

Elementos de la orientación a objetos:
1. Clases. 
2. Propiedades
3. Métodos
4. Objetos

Conceptos claves de POO
- Encapsulamiento: Hacer que un dato, un atributo, un método o incluso una clase sea inviolabe, inalterable. se le asigna un modificador de acceso.
- Abstracción: Cuando nosotros abstraemos los datos de un objeto para generar un molde.
- Herencia
- Polimorfismo: Capacidad que tienen ciertos lenguajes para hacer que, al enviar el mismo mensaje (Invocar el mismo metodo) desde distintos objetos, cada uno de esos objetos pueda responder a ese mensaje (Invocacion) de forma distinta. En resumen, construir métodos con el mismo nombre pero con comportamiento diferente.

##Que es un objeto?

Cuando tengamos un problema lo primero que debemos hacer es identificar objetos. 

Los objetos siempre van a tener:
- Propiedades (Tambien llamadas atributos) y comportamientos.
- Son sustantivos. 
- Pueden ser fisicos (Manzana, persona) o conceptuales (Pueden ser simbologias de la mente --> Una sesión de usuario, la cuenta de un banco).

**Propiedades/Atributos**

Los atributos tambien seran sustantivos. Por ejemplo: Nombre, tamaño, forma, estado (todo lo que compone las caracteristicas de ese objeto). 

**Comportamientos**

Son todas las operaciones del objeto. Suelen ser verbos o sustantivo y verbo (makeReport).

## Que es una clase?

Es el modelo sobre el cual se construira nuestro objeto. Conectando con el ejemplo anterior, es construir la forma mas general de un perrito (Para permitir generar mas perritos). Es similar a crear un molde. En resumen, las clases son los modelos sobre los cuales construimos objetos.

Abstracción: Cuando nosotros abstraemos los datos de un objeto para generar un molde.

Las clases permiten:
- La modularidad.
- Dividir el programa en diferentes partes o modulos.
- Separar las clases en archivos (las clases deben vivir en archivos diferentes). --> Asi se mantiene la modularidad.

## Modularidad

Subdividir un sistema en partes mas pequeñas (Modulos). Estos modulos trabajan de forma independiente para componer un sistema completo. Esto es una caracteristica que diferencia mucho la POO de la programación estructurada.

Ventajas de la modularidad:
- **Reutilizar** codigo.
- **Evitar un colapso** masivo en nuestro código.
- Mejorar la **legibilidad**.
- Facilitar el debug y mantenimiento.Cuando hagas una actualización, lo podrías hacer por módulo y no causar cosas inesperadas en tu sistema (Resolución rapida de problemas).

## Aplicando Codigo

### Generación de objetos

Los objetos nos ayudan a crear instancia de una clase, el objeto es el resultado de lo que modelamos, de los parámetros declarados y usaremos los objetos para que nuestras clases cobren vida.

Los métodos constructores dan un estado inicial al objeto y podemos añadirle algunos datos al objeto mediante estos métodos. Los atributos o elementos que pasemos a través del constructor serán los datos mínimos que necesita el objeto para que pueda vivir. El método constructor es lo que va despues del "=". Por ejemplo en Python seria la parte de la sentencia "Person()"

En Python encontrarás un concepto denominado **Métodos Mágicos**, estos métodos son llamados automática y estrictamente bajo ciertas reglas. El método constructor en Python forma parte de esta familia de métodos y como aprendimos en la clase anterior lo declaramos usando __init__, aunque si nos ponemos estrictos este método no construye el objeto en sí. El encargado de hacer esto es __new__ y el método __init__ se encargará de personalizar la instanciación de la clase, esto significa que lo que esté dentro de __init__ será lo primero que se ejecute cuando se cree un objeto de esta clase.


Para nuestro proyecto tenemos la necesidad de que algunas variables se inicialicen obligatoriamente cuando ocurra la instanciación. Así que declaremos el método __init__ en las clases de nuestro proyecto con las propiedades obligatorias.


Para la clase Account quedaría algo así, notarás que usamos la palabra clave self, esta es muy parecida a lo que venimos trabajando a otros lenguajes con this. Y como su nombre lo dice hace referencia a los datos que componen la clase, en este caso self.name está llamando al atributo name que se encuentra en la línea 3 de la clase y, le está asignando el dato que se pasa en el método __init__ de la línea 8.
