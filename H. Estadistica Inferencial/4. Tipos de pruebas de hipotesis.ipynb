```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as st
from scipy.stats import f_oneway
from scipy.stats import ttest_ind
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
iris = pd.read_csv(url, names = names)
iris.head()
```

### Tipos de pruebas

### a) Distribucion t de student

image.png

```python
se_length = iris.std()['sepal-length']/np.sqrt(10)
se_width = iris.std()['sepal-width']/np.sqrt(10)
```

image.png

```python
sed = np.sqrt((se_length**2)+(se_width**2))
```

image.png

```python
t_stat = (iris.mean()['sepal-length']-iris.mean()['sepal-width'])/sed
t_stat
```

```python
p = ttest_ind(iris['sepal-length'], iris['sepal-width'])
p
```

```python
" Se rechaza que la distribucion de sepal-length sea igual a la distribucion de sepal-width. El valor de T es mucho mayor al p-value"
```

### b) Correlacion y coeficiente de Pearson

```python
iris['sepal-length'].corr(iris['sepal-width'])
```

```python
st.pearsonr(iris['sepal-length'], iris['sepal-width'])
```

###c) Analisis de la varianza ANOVA

```python
f_oneway(iris['sepal-length'], iris['sepal-width'])
```
