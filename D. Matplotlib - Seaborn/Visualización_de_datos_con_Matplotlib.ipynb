##2.Subplot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.subplot(1,2,1) #1er argumento: # de filas de graficos; 2do argumento: # de columnas de graficos; 3er argumento: index entre filas y columna (Para este caso se crearian 2 index porque son 2 graficas)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.hist(y)
plt.show
```

```python
plt.subplot(1,2,1) 
plt.plot(x,y,'r--')
plt.plot(y,x,'b:')
plt.subplot(1,2,2)
plt.pie(y)
plt.show()
```

```python
plt.subplot(2,1,1) 
plt.plot(x,y,'r--')
plt.plot(y,x,'b:')
plt.subplot(2,1,2)
plt.pie(y)
plt.show()
```

```python
#EJEMPLO DE COMPA√ëERO QUE ME LLAMO LA ATENCI√ìN - NO ES DE LA CLASE

import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec

fig = plt.figure(tight_layout=True)
gs = gridspec.GridSpec(2, 2) # 2 filas x 2 columnas 
ax = fig.add_subplot(gs[0, :]) # comparto toda la fila 0
ax.hist(y) 
ax = fig.add_subplot(gs[1, 0])
ax.plot(x,y,'r--')  
ax = fig.add_subplot(gs[1, 1])
ax.plot(y,x,'b:') 
plt.show()
```

##3.M√©todo orientado a objetos

image.png

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

image.png

- Figure: Es el linzo.
- Axes: Es la grafica.
- Axis: Son los ejes de la grafica.

```python
fig = plt.figure() #En el m√©todo orientado a objetos primero se crea un lienzo con esta sentencia. 
axes = fig.add_axes([0.1,0.1,0.2,0.9]) #Los primeros dos parametros "0.1,0.1" hacen referencia a la posici√≥n en el lienzo. Los siguientes dos hacen referencia al tama√±o que va a tener la figura en ese lienzo. 
#Tener en cuenta que cuando solo es una figura, los primeros dos parametros, no hacen efecto. Sin embargo para dos figuras si que sirven como se ver√° en el siguiente ejemplo
axes.plot(x,y,'b')
fig.show()
```

```python
fig = plt.figure() #En el m√©todo orientado a objetos primero se crea un lienzo con esta sentencia. 
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.1,0.1,0.4,0.3])

axes.plot(x,y,'b')
axes2.plot(y,x,'r')
fig.show()
```

```python
fig = plt.figure() #En el m√©todo orientado a objetos primero se crea un lienzo con esta sentencia. 
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.2,0.6,0.4,0.3])

axes.plot(x,y,'b')
axes2.plot(y,x,'r')
axes2.set_facecolor('gray')
fig.show()
```

## 4.Subplots

La diferencia entres "subplot" y "subplots" es que la 2da es para plotear varias graficas pero en el m√©todo orientado a objetos. 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x,y,'b')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
#axes.plot(x,y,'b')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'b') #axes[0]-->Hace referencia al primer grafico, el de indice 0. 
axes[1].plot(y,x,'r')
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2) #Tambien puedo definir una tupla "(ax1,ax2)" pero obvio debe coincidir con la cantidad de graficos que defina en nrows y ncols
ax1.plot(x,y,'b') 
ax2.plot(y,x,'r')
```

```python
fig, axes = plt.subplots(2, 4) #Poner nrows y ncols es opcional
axes[0,0].plot(x,np.cos(x),'b') 
axes[0,1].plot(x,np.sin(x),'y') 
axes[0,2].plot(x,np.tan(x),'r') 
axes[0,3].plot(x,np.cos(x)**2,'b') 
```

```python
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(2, 4) #Poner nrows y ncols es opcional
ax1.plot(x,np.cos(x),'b') 
ax2.plot(x,np.sin(x),'y') 
ax3.plot(x,np.tan(x),'r') 
ax4.plot(x,np.cos(x)**2,'b') 
fig.tight_layout() # Para que se organicen de mejor manera las graficas
```

##5.Leyendas, etiquetas, titulos y Tama√±o

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(1,2, figsize=(10,5))
axes[0].plot(x,y, label="sin(x)") # Defino los labels (Leyenda)en la grafica y despues debo ejecutar "axes[0].legend()"
axes[0].set_title('Relacion X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()

axes[1].plot(y,x)
axes[1].set_title('Relacion Y - X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

```python
#PARA Pyplot Basico
plt.plot(x,y,label="$sin(x)$") #El simbolo de pesos hace que la leyenda use notaci√≥n matematica
plt.title("Este es un titulo")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc='lower left', bbox_to_anchor=(1,0.9))
```

### 6.2 Estilo Pyplot

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1, color = 'green', alpha = 1, linewidth=5, linestyle="-", marker='o', markersize = 10, markerfacecolor='#CCFFCC') #alpha: Transparencia del color
ax.plot(x,x+2, color = 'blue', linewidth=3, linestyle="--", marker='x', markersize = 4)
ax.plot(x,x+3, color = '#66FF66', linewidth=8, linestyle="dashed", marker='v') #RGB
ax.plot(x,x+4, color = '#CCFFCC', linewidth=1, linestyle=":", marker='P')
```

##7.Bar Plot

## 8.Otros tipos de graficos

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100)
```

```python
data
```

###8.1Histograma

```python
plt.hist(data, bins=50, width=1) #bins --> Numero de barras
plt.show()
```

```python
plt.hist(data, bins=50, histtype='step') #bins --> Numero de barras
plt.show()
```

###8.2 Boxplot

```python
data = np.append(data, 200)
plt.boxplot(data, vert=False, patch_artist=True, notch = True, showfliers=False) #notch --> Hace el embudito; patch_artist --> Para rellenar el grafico de boxplot
#showfliers=False: Hace que no me muestre los outliers
plt.show()
```

### 8.3 Grafico de dispersi√≥n

```python
N=50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30*np.random.rand(N))**2
colors = np.random.rand(N)
```

```python
plt.scatter(x,y, s=area, c=colors, marker='p', alpha=0.5)
plt.show()
```
