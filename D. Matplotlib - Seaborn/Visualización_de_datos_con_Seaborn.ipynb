##4.Distribuciones

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips4 = sns.load_dataset('tips')
tips4.head()
```

### 4.1Histplot (Histograma)

```python
sns.histplot(data=tips4, x='tip', bins=15)
plt.show()
```

```python
sns.histplot(data=tips4, x='tip', bins=15, cumulative=True)
plt.show()
```

```python
sns.histplot(data=tips4, x='tip', bins=15, hue='sex', stat='frequency') #Calcula el estadístico de la frecuencia. El default es el count. 
plt.show()
```

```python
sns.histplot(data=tips4, x='tip', bins=15, hue='sex', stat='probability') #Calcula el estadístico de la probabilidad
plt.show()
```

```python
sns.histplot(data=tips4, x='tip', bins=15, hue='sex', stat='count', multiple='dodge') 
#Parametro multiple --> stack:Acumula la categoria, dodge: Pone una categoria al lado de la otra;layer: Una categoria detras de la otra (Default); fill: De 0 a 1 por categoria
plt.show()
```

### 4.2kdeplot (Linea)

```python
sns.kdeplot(data=tips4, x="tip", hue="sex", cumulative = False, shade=True) #Shade: Pinta el area bajo la curva
```

###4.3ecdfplot (Acumulativo)

```python
sns.ecdfplot(data=tips4, x='tip', hue='sex', stat='count')
plt.show()
```

### 4.4 displot (Hta Gral Dist)

Usando el parametro "kind" puedo cambiar el tipo de grafico. Por lo que puede hacer lo mismo que se grafico en 4.1, 4.2 y 4.3

```python
sns.displot(data=tips4, x='tip', hue='sex', kind='kde') #kind
plt.show()
```

##5.Graficos Categoricos

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips5 = sns.load_dataset('tips')
tips5.head()
```

###5.1 Countplot

```python
sns.countplot(data=tips5, x='day', hue='sex')
plt.show()
```

```python
sns.countplot(data=tips5, y='day', hue='sex')
plt.show()
```

### 5.2 Stripplot

```python
plt.figure(figsize=(6,6))
sns.stripplot(data=tips5, x='day', y='total_bill', hue='sex',dodge=True)
plt.show()
```

###5.3 Swarmplot

```python
plt.figure(figsize=(6,6))
sns.swarmplot(data=tips5, x='day', y='total_bill', hue='sex',dodge=True)
plt.show()
```

###5.4 Boxplot

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips5, x='day', y='total_bill', hue='sex',dodge=True)
plt.show()
```

### 5.5 Violin

```python
plt.figure(figsize=(6,6))
sns.violinplot(data=tips5, x='day', y='total_bill', hue='sex',dodge=True, palette='pastel')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.violinplot(data=tips5, x='day', y='total_bill', hue='sex',dodge=True, palette='pastel', split=True) #Utilizando split=True
plt.show()
```

###5.6 Catplot

Usando el parametro "kind" puedo cambiar el tipo de grafico. Por lo que puede hacer lo mismo que se grafico en 5.1, 5.2, 5.3, 5.4, 5.5. Es muy similar a "displot" pero para graficos categoricos.

```python
plt.figure(figsize=(6,6))
sns.catplot(data=tips5, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.catplot(data=tips5, x='day', y='total_bill', hue='sex', dodge=True, kind="swarm")
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.catplot(data=tips5, x='day', y='total_bill', hue='sex', dodge=True, kind="box", col='time')
plt.show()
```

###5.7 Combinando

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips5, x='day', y='total_bill', hue='sex',dodge=True)
sns.swarmplot(data=tips5, x='day', y='total_bill', hue='sex',dodge=True, color='0', marker='<')
plt.show()
```

##6.Graficos de Relación

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips6 = sns.load_dataset('tips')
tips6.head()
```

###6.1 Scatterplot

```python
plt.figure(figsize=(6,6))
sns.scatterplot(data=tips6,x='total_bill', y='tip', hue='day', style='time', size='size')
plt.legend(loc='center', bbox_to_anchor=(1.12,0.6))
plt.show()
```

```python
plt.figure(figsize=(8,8))
markers = {"Lunch":"D", "Dinner":"s"} #Para definir como quiero los markers para la categoria
sns.scatterplot(data=tips6,x='total_bill', y='tip', hue='day', style='time', size='size', markers=markers)
plt.legend(loc='center', bbox_to_anchor=(1.12,0.6))
plt.show()
```

### 6.2 Lineplot

```python
sns.lineplot(data=tips6,x='total_bill', y='tip', hue='day', style='time', size='size', markers=markers)
plt.show()
```

No muy optimo. Pero podria funcionar para pocas cateorias. Sin embargo, mucho mejor el scatterplot. N

###6.3 Relplot

Es el simil de "Catplot" y "displot" pero para graficas de relación.

```python
sns.relplot(data=tips6,x='total_bill', y='tip', hue='day', style='time', size='size', markers=markers, col="time")
plt.show()
```

## 7.Joinplot-Pairplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips7 = sns.load_dataset('tips')
tips7.head()
```

###7.1JointPlot

```python
sns.jointplot(data=tips7, x='total_bill', y='tip', hue='sex', marginal_ticks=True, kind='hist', marginal_kws=dict(bins=25, fill=False, multiple='dodge')) 
#marginal_ticks: Ejes para los graficos auxiliares. 
plt.show()
```

###7.2 Pairplot

Para mirar las relaciones entre variables numericas. Automaticamente detecta cuales son dichas variables numericas en el dataFrame

```python
sns.pairplot(data=tips7, hue='sex', palette='coolwarm', corner=True, kind='scatter') #Corner--> Para no tener graficos duplicados.
plt.show()
```

##8.Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips8 = sns.load_dataset('tips')
tips8.head()
```

```python
tips8.corr()
```

```python
sns.heatmap(tips8.corr())
plt.show()
```

```python
sns.heatmap(tips8.corr(), annot=True, cmap='coolwarm', linewidths=5, linecolor= "black", vmin=0, vmax=1, cbar=True) 
#annot: para que aparesca el valor en cada recuadro. 
#vmin y vmax= Es para modificar los maximos y minimos de la paleta de colores. 
#cbar: Quitar o poner la barra de la derecha
plt.show()
```

```python
corr_matrix = tips8.corr()
```

```python
corr_matrix.unstack()
```

```python
corr_values=corr_matrix.unstack()
```

```python
corr_values.sort_values(kind='quicksort')
```

Si se cambia en cmap el valor de “coolwarm” agregandole “_r” se invierte el sentido de los colores
